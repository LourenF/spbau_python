# Гиперболический косинус

$$cosh x = 1 + \frac{x^2}{2!} + \frac{x^4}{4!} + \frac{x^6}{6!} + \ldots + R(x)=
\sum_{n=0}^N \frac{x^{2n}}{2n!} + R(x)$$

$$R(x) \xrightarrow[N \rightarrow \infty]{} 0$$

```python
import math

ITERATIONS = 20

def my_cosh(x):
    """
    Вычисление гиперболического косинуса при помощи частичного суммирования
    ряда Тейлора для окрестности 0
    """
    x_pow = 1
    multiplier = 1
    partial_sum = 1
    for n in range(1, ITERATIONS):
        x_pow *= x**2  # В цикле постепенно считаем степень
        multiplier *= 1 / (2*n) / (2*n - 1) # и факториал
        partial_sum += x_pow * multiplier
    
    return partial_sum

print(help(math.cosh), math.cosh(0.3))
print(help(my_cosh), my_cosh(0.3))
```

```python
import math
import cmath

complex_angle = cmath.acosh(5)
print('"Угол", на котором гиперболический косинус достигает пяти:', complex_angle)
print('Библиотечный гиперболический косинус достигает пяти:', cmath.cosh(complex_angle))
print('Наш гиперболический косинус достигает пяти:', my_cosh(complex_angle))
```

```python
%matplotlib inline
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('svg')

import matplotlib.pyplot as plt
import numpy as np

vs = np.vectorize(my_cosh)

angles = np.r_[-30.25:30.25:0.01]
plt.plot(angles, np.cosh(angles))
plt.plot(angles, vs(angles))
plt.show()
```
